## Image Classification with ResNet50

```python
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array, array_to_img, load_img
from tensorflow.keras.applications.resnet50 import ResNet50, decode_predictions, preprocess_input
```

### - Load image and Preprocess
- Images can be loaded using load_img function

```python
img = load_img('dog.jpg', target_size=(224,224))   ##image size can be calibrated with target_size parameter 
                                                    ##(target_size shoudl be (224,224) for ResNet50)
img
```

```python
img_arr = img_to_array(img)

print(img_arr.shape)
```

```python
img_arr = np.expand_dims(img_arr, axis=0)

print(img_arr.shape)
```

### - Model Predict

```python
model = ResNet50(weights='imagenet')
```

```python
pred_class = model.predict(img_arr)
```

```python
# print only top 10 predicted classes
n = 10
top_n = decode_predictions(pred_class, top=n)
```

```python
for c in top_n[0]:
    print(c)
```

##### - Preprocess image with preprocess_input() of Model and Predict Again

```python
img_processed = preprocess_input(img_arr)
```

```python
pred_class = model.predict(img_processed)
```

```python
# print only top 10 predicted classes
n = 10
top_n = decode_predictions(pred_class, top=n)
```

```python
for c in top_n[0]:
    print(c)
```
